### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
above_median_mpg_cars = df[df['mpg'] > median_mpg].copy()
above_median_mpg_cars['hp_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']
best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight_ratio'] == above_median_mpg_cars['hp_to_weight_ratio'].max()]
print("\nCar(s) with the best ratio of horsepower to weight among above-median mpg cars:")
print(best_ratio_cars[['name', 'mpg', 'horsepower', 'weight', 'hp_to_weight_ratio']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(12, 6))
avg_mpg_origin_year = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()
avg_mpg_origin_year.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(df['model_year'].unique())
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
car_years_count = df.groupby('name')['model_year'].nunique()

multi_year_car_names = car_years_count[car_years_count >= 2].index

multi_year_cars_df = df[df['name'].isin(multi_year_car_names)].copy()

car_mpg_stats = multi_year_cars_df.groupby('name').agg(
    std_mpg=('mpg', 'std'),
    avg_mpg=('mpg', 'mean'),
    appearances=('name', 'count')
).dropna()

consistent_cars_final = car_mpg_stats[car_mpg_stats['std_mpg'] < 1.0]

final_sorted_consistent_cars = consistent_cars_final.sort_values(
    by=['appearances', 'avg_mpg'],
    ascending=[False, False]
)

print("Consistent Car Models (low variation in mpg, produced across multiple years):")
print(final_sorted_consistent_cars[['appearances', 'avg_mpg', 'std_mpg']])
```
